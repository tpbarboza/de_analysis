### Use of internationalization table example

```python
sql = """select
            p.id,
            c."name" category,
            i."text" as category_english,
            p.name_length,
            p.description_length,
            p.photos_qty,
            p.weight_g,
            p.length_cm,
            p.height_cm,
            p.width_cm
        from product as p
        join category c on c.id = p.category_id
        join internationalization i on i.reference_id = c.id and i.reference_table = 'category'
        join "language" l on l.id = i.language_id and l.code = 'EN-US'"""

df_prod = pd.read_sql(sql, conn)
df_prod
```

### TOP 10 sold products 

```python
sql = """with top_sold as (select product_id, count(1) as total from sales_order_item as oi
group by product_id order by 2 desc limit 10) 
select cat.name as category, p.* from product p 
join top_sold on p.id = top_sold.product_id
join category cat on cat.id = p.category_id"""
df_top_sold = pd.read_sql(sql, conn)
df_top_sold.profile_report()
```

### Best month for top product sales

```python
sql = """with top_sold_date as (
with top_sold as (select product_id, count(1) as total from sales_order_item as oi
group by product_id order by 2 desc limit 10) 
select oi.product_id, to_char(o.purchase_timestamp, 'YYYY-MM') as purchase_month, count(1) as sold_qty
from sales_order o
join sales_order_item oi on o.id = oi.sales_order_id
join top_sold on top_sold.product_id = oi.product_id
group by oi.product_id, purchase_month, top_sold.total
order by product_id, sold_qty desc)
select product_id, purchase_month, cat."name", sold_qty
from top_sold_date tsd
join product p on p.id = tsd.product_id
join category cat on cat.id = p.category_id
where sold_qty = (select max(sold_qty) from top_sold_date where product_id = tsd.product_id)
order by sold_qty desc"""
df_top_sold_date_for_top_sold_product = pd.read_sql(sql, conn)
df_top_sold_date_for_top_sold_product
```

### Worst month for top product sales

```python
sql = """with top_sold_date as (
with top_sold as (select product_id, count(1) as total from sales_order_item as oi
group by product_id order by 2 desc limit 10) 
select oi.product_id, to_char(o.purchase_timestamp, 'YYYY-MM') as purchase_month, count(1) as sold_qty
from sales_order o
join sales_order_item oi on o.id = oi.sales_order_id
join top_sold on top_sold.product_id = oi.product_id
group by oi.product_id, purchase_month, top_sold.total
order by product_id, sold_qty asc)
select product_id, purchase_month, cat."name", sold_qty
from top_sold_date tsd
join product p on p.id = tsd.product_id
join category cat on cat.id = p.category_id
where sold_qty = (select min(sold_qty) from top_sold_date where product_id = tsd.product_id)
order by sold_qty desc"""
df_worst_date_for_top_sold_product = pd.read_sql(sql, conn)
df_worst_date_for_top_sold_product
```

### Lets take a sample of 4 months for the 50 top seller products, and verify if there are great price variance among them within the same month

```python
sql = """with top_sold as (select product_id, count(1) as total from sales_order_item as oi
group by product_id order by 2 desc limit 50) 
select oi.product_id, to_char(purchase_timestamp, 'YYYY-MM') purchase_month, max(price), min(price), (min(price) / max(price)) * 100 as percentage 
from sales_order_item as oi
join sales_order o on o.id = oi.sales_order_id
join top_sold ts on ts.product_id = oi.product_id
where o.purchase_timestamp between '2018-04-01' and '2018-08-31'
group by oi.product_id,  purchase_month
having max(price ) != min(price)
order by percentage asc, product_id desc, purchase_month desc
"""
df_prod_price_diff = pd.read_sql(sql, conn)
df_prod_price_diff.head(20)
```

### As we can see, for the top selling products, there are just 10 cases where price different is "expressive" (above 15%). Lets check for all products for same period: 

```python
sql = """with t as (select oi.product_id, to_char(purchase_timestamp, 'YYYY-MM') purchase_month, max(price), min(price), (min(price) / max(price)) * 100 as percentage 
from sales_order_item as oi
join sales_order o on o.id = oi.sales_order_id
where o.purchase_timestamp between '2018-04-01' and '2018-08-31'
group by oi.product_id,  purchase_month
having max(price ) != min(price)
order by percentage asc, product_id desc, purchase_month desc)
select product_id, purchase_month, percentage
from t
where percentage < 85
order by 2 desc
"""
df_all_prod_price_diff = pd.read_sql(sql, conn)
df_all_prod_price_diff.count()
```

```python
sql = """select count(distinct product_id )
from sales_order_item as oi
join sales_order as o on oi.sales_order_id = o.id
where o.purchase_timestamp between '2018-04-01' and '2018-08-31'
"""
df_total_prod_sold = pd.read_sql(sql, conn)
df_total_prod_sold
```

### For the same period, a total of 14696 distinc products were sold, but only 140 (~0.01%) had a price difference above 15%. I'd say that the pricing is being kept average among sellers.

### Lets check if product description, name lenght of photo quantity play an decisive role between top sold and bottom sold products

```python
sql = """with top_sold as (select product_id, count(1) as total from sales_order_item as oi
group by product_id order by 2 desc limit 10) 
select avg(p.description_length) desc_len, avg(p.name_length) name_len, avg(p.photos_qty) pht_qty
from product p 
join top_sold on p.id = top_sold.product_id
join category cat on cat.id = p.category_id"""
df_best_sell_info = pd.read_sql(sql, conn)
df_best_sell_info
```

```python
sql = """with worst_sold as (select product_id, count(1) as total from sales_order_item as oi
group by product_id order by 2 asc limit 10) 
select avg(p.description_length) desc_len, avg(p.name_length) name_len, avg(p.photos_qty) pht_qty
from product p 
join worst_sold on p.id = worst_sold.product_id
join category cat on cat.id = p.category_id"""
df_worst_sell_info = pd.read_sql(sql, conn)
df_worst_sell_info
```

### As shown above the difference between the average lenght of description, name and photos quantity is virtually the same amongst the best and worst sold products (actually, a bit "better" in the worst sold), thus not as decisive as decision maker. Might look odd, but data don't lie.

### LETS EVALUATE FREIGTH. Lets check the main regions for sellers and customers, and use that as strategy.

```python
sql = """with region as (select "name" as state_name, case 
	when "name" in ('SP', 'RJ', 'MG', 'ES') then 'SUDESTE'
	when "name" in ('AC','AP','AM','PA','RO','RR','TO') then 'NORTE'
	when "name" in ('MA','PI','CE','RN','PB','PE','AL','SE','BA') then 'NORDESTE'
	when "name" in ('DF','GO','MT','MS') then 'CENTRO-OESTE'
	when "name" in ('PR','SC','RS') then 'SUL'
END as region_name from state)
select region.region_name, count(1)
from seller s
join zip on s.zip_id = zip.id
join city on zip.city_id = city.id
join state on city.state_id = state.id
join region on state."name" = region.state_name
group by region.region_name
order by 2 desc;
"""
df_seller_by_region = pd.read_sql(sql, conn)
df_seller_by_region
```

```python
sql = """with region as (select "name" as state_name, case 
	when "name" in ('SP', 'RJ', 'MG', 'ES') then 'SUDESTE'
	when "name" in ('AC','AP','AM','PA','RO','RR','TO') then 'NORTE'
	when "name" in ('MA','PI','CE','RN','PB','PE','AL','SE','BA') then 'NORDESTE'
	when "name" in ('DF','GO','MT','MS') then 'CENTRO-OESTE'
	when "name" in ('PR','SC','RS') then 'SUL'
END as region_name from state)
select region.region_name, count(1)
from customer c
join zip on c.zip_id = zip.id
join city on zip.city_id = city.id
join state on city.state_id = state.id
join region on state."name" = region.state_name
group by region.region_name
order by 2 desc;
"""
df_customer_by_region = pd.read_sql(sql, conn)
df_customer_by_region
```

### Average freigth value among regions

```python
sql = """with region as (select "name" as state_name, case 
	when "name" in ('SP', 'RJ', 'MG', 'ES') then 'SUDESTE'
	when "name" in ('AC','AP','AM','PA','RO','RR','TO') then 'NORTE'
	when "name" in ('MA','PI','CE','RN','PB','PE','AL','SE','BA') then 'NORDESTE'
	when "name" in ('DF','GO','MT','MS') then 'CENTRO-OESTE'
	when "name" in ('PR','SC','RS') then 'SUL'
END as region_name from state)
select sell_reg.region_name as seller_region,
		cust_reg.region_name as customer_region,		
		count(1) as transactions,
		stddev(oi.freight_value) stddev_freigth
from sales_order o
join sales_order_item oi on o.id = oi.sales_order_id
join customer c on c.id = o.customer_id
join seller s on s.id = oi.seller_id
join zip as cust_zip on cust_zip.id = c.zip_id
join city as cust_cit on cust_cit.id = cust_zip.city_id
join state cust_st on cust_st.id = cust_st.id
join region as cust_reg on cust_reg.state_name = cust_st."name"
join zip as sell_zip on sell_zip.id = s.zip_id
join city as sell_cit on sell_cit.id = sell_zip.city_id
join state as sell_st on sell_st.id = sell_cit.state_id
join region as sell_reg on sell_reg.state_name = sell_st."name"
group by customer_region, seller_region
order by stddev_freigth desc
"""
df_avg_freigth = pd.read_sql(sql, conn)
df_avg_freigth
```

**As observed, the standard deviation of the freight values amont regions shows that it is cheaper to ship from the center region, and most expensive to ship from the south region. We also notice that the deviation is almost the same within the same origin region.**

**Lets reorder the dataframe to see what are the top transactions**

```python
df_avg_freigth.sort_values('transactions', ascending=False)
```

**As we saw in previous datasets, the most sellers are from southeast and south, and in fact these are the regions which makes most of the sells., However, although we also saw that the most customers are from southeast and south regions as well, the table above shows us that actually the top customers are from northeast, north and center regions.
Crossing on all this information it is possible to think of scenarios of incentives, for both sellers and customers.
